```python
import pandas as pd
import numpy as np
import scipy

pd.options.display.max_rows = 1000
pd.options.display.max_columns = 1000
pd.options.display.max_colwidth = 999
```

```python
def get_method_type(x):
    if "RevKD" in x:
        return "RevKD"
    elif "GKD" in x:
        return "GKD"
    else:
        return "SeqKD"
```

```python
data = pd.read_csv("../cs_results_main.csv")
```

```python
data2 = pd.read_csv("../llama_cs_results.csv")
```

```python
data = pd.concat([data, data2], axis=0)
```

```python
data = data.rename(columns={'hellaswag':'Hellaswag','boolq':'BoolQ','social_i_qa':'SiQA','piqa':'PiQA','ARC-Easy':'ARC-e', \
                               'ARC-Challenge':'ARC-c','winogrande':'Winogrande','openbookqa':'OBQA'})
```

```python
data1 = data[data.Method == 'SFT']
data2 = data[data.Method.str.contains("student")]
#data2 = data2[data2.Method.str.contains("temp_1")]

data2['Method'] = data2['Method'].apply(get_method_type)
```

```python
data2['Teacher'] = data2['Teacher'].apply(lambda x: x.split("_")[0])
```

```python
data2
```

```python
tasks = data.columns[3:-1].tolist()
tasks.sort()

tasks += ['Average']
```

```python
data2 = data2.groupby(['Teacher','Student','Method'])[tasks].max().reset_index(drop=False)
```

```python
#data2 = data2.groupby(['Teacher','Student'])[tasks].max().reset_index(drop=False)
```

### Hypothesis 1

Statistical test of KD vs SFT improvement

```python
data3 = pd.merge(data2[['Teacher','Student','Method'] + tasks], \
                 data1[['Student'] + tasks].rename(columns={task: task+" SFT" for task in tasks}), how='inner')
```

```python
for task in tasks:
    data3[f'{task} Improvement'] = data3[task] - data3[f"{task} SFT"]
```

```python
write_df = pd.DataFrame()
idx = 0
```

```python
for task in tasks:
    ttest = scipy.stats.ttest_1samp(data3[f'{task} Improvement'], 0, alternative='greater')
    print (task, ttest)
    write_df.loc[idx, "Test Type"] = "t-test all methods"
    write_df.loc[idx, task] = "{} ({})".format(round(ttest.statistic,2), round(ttest.pvalue, 2))
```

```python
idx += 1

for method in data3.Method.unique():
    for task in tasks:
        ttest = scipy.stats.ttest_1samp(data3[data3.Method == method][f'{task} Improvement'], 0, alternative='greater')
        write_df.loc[idx, "Test Type"] = "t-test {}".format(method)
        write_df.loc[idx, task] = "{} ({})".format(round(ttest.statistic,2), round(ttest.pvalue, 2))
        print (method, task, ttest)
    
    idx += 1
        
        
```

```python
for task in tasks:
    a = data3[data3.Method == 'SeqKD'][f'{task} Improvement'].values
    b = data3[data3.Method == 'RevKD'][f'{task} Improvement'].values
    c = data3[data3.Method == 'GKD'][f'{task} Improvement'].values

    F = scipy.stats.f_oneway(a, b, c)
    print (task, F)
    write_df.loc[idx, "Test Type"] = "ANOVA all methods"
    write_df.loc[idx, task] = "{} ({})".format(round(F.statistic,2), round(F.pvalue, 2))
```

```python
write_df.to_csv("hypo1_cs.csv",index=False)
write_df
```
